Par ou ímpar

```python
#Entrada
número = int (input("Digite um número de 0 até 20:"))

#Processamento&Saída

if número %2 == 0:
    print(f"{número} Seu número é Par")

else:
    print(f"Seu número é Ímpar")
```

Aprovação de aluno

```python
#Aprovação

nota = int(input("Digite sua nota:"))

if nota >= 7:
    print("Você foi Aprovado!!")
elif nota >= 5 and nota < 7:
    print("Você está de Recuperação!!")
else:
    print("Você está Reprovado!!")
```

Classificação de idade

```python
#Idade

idade = int (input("Digite sua idade:"))

if 0 <= idade <= 12:
    print("Você é Criança!!")
elif 13 <= idade <= 17:
    print("Você é Adolescente!!")
elif 18 <= idade <= 59:
    print("Você é Adulto")
else:
    print("O(A) senhor(a) é Idoso(a)")
```

Positivo, negativo ou zero

```python
#Negativo, Positivo ou Zero

numero = float(input("Digite um número:"))

if numero == 0: print("Seu número é Zero!")

elif numero < 0: print("Seu número é Negativo!")

else:
  print("Seu número é Positivo!")
```

Calculadora simples

```python
#Calculadora

numero1 = int(input("Digite um número:"))
numero2 = int(input("Digite outro número:"))
operacao = str(input("Escolha uma operação(+,-,*,/): "))

resultado1 = numero1 + numero2
resultado2 = numero1 - numero2
resultado3 = numero1 * numero2
resultado4 = numero1 / numero2

if operacao == "+":
    print(f"Seu resultado é {resultado1}!")
elif operacao == "-":
    print(f"Seu resultado é {resultado2}!")
elif operacao == "*":
    print(f"Seu resultado é {resultado3}!")
elif operacao == "/":
    print(f"Seu resultado é {resultado4}!")
else:
    print("Operação inválida!")
```

Dia da semana

```python
#Semana

dia = int(input("Digite um número de 1 a 7 correspondente ao dia da semana:"))

if dia == 1: print("O dia da semana é Domingo!")

elif dia == 2: print("O dia da semana é Segunda-feira!")

elif dia == 3: print("O dia da semana é Terça-feira!")

elif dia == 4: print("O dia da semana é Quarta-feira!")

elif dia == 5: print("O dia da semana é Quinta-feira!")

elif dia == 6: print("O dia da semana é Sexta-feira!")

elif dia == 7: print("O dia da semana é Sábado!")

else:
  print("Dia inválido!")
```

Ano bissexto

```python
#Ano Bissexto

ano = int(input("Digite um ano:"))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0): print("O ano é bissexto!")

else:
  print("O ano não é bissexto!")
```
