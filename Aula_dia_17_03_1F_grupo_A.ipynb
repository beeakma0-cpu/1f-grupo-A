```python
# Operadores aritméticos
# Fazer um programa que caucula com 02 números

# Entrada

a = int(input("Digite um número >;): "))
b = int(input("Digite um número >;): "))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_Int = a // b
resto = a % b
potencia = a ** b

# Saída
print("")
print("~~~~~~~~~~OPERADORES ARITMÉTICOS~~~~~~~~~~")
print("")
print("~~~~~~~~~~~~~~~~~~SOMA~~~~~~~~~~~~~~~~~~")
print("")
print(f" O resultado da soma de {a} mais {b} é igual a: {soma} <3")
print("")
print("~~~~~~~~~~~~~~~~SUBTRAÇÃO~~~~~~~~~~~~~~~~")
print("")
print(f" O resultado da subtração de {a} menos {b} é igual a: {subtracao} <3")
print("")
print("~~~~~~~~~~MULTIPLICAÇÃO~~~~~~~~~~")
print("")
print(f" O resultado da multiplicação de {a} vezes {b} é igual a: {multiplicacao} <3")
print("")
print("~~~~~~~~~~~~~~~~DIVISÃO~~~~~~~~~~~~~~~~")
print("")
print(f" O resultado da divisão de {a} dividido por {b} é igual a: {divisao} <3")
print("")
print("~~~~~~~~~~DIVISÃO INTEIRA~~~~~~~~~~")
print("")
print(f" O resultado da divisão inteira de {a} dividido por {b} é igual a: {div_Int} <3")
print("")
print("~~~~~~~~~~RESTO DA DIVISÃO~~~~~~~~~~")
print("")
print(f" O resultado da divisão inteira de {a} dividido por {b} é o resto da divisão é igual a: {resto} <3")
print("")
print("~~~~~~~~~~POTÊNCIAÇÃO~~~~~~~~~~")
print("")
print(f" O resultado da potenciação de {a} elevado {b} é igual a: {potencia} <3")
```
